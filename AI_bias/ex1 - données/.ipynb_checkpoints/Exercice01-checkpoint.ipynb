```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
X_train = np.load('regression_xtrain.npy')
y_train = np.load('regression_ytrain.npy')
X_test = np.load('regression_xtest.npy')
y_test = np.load('regression_ytest.npy')

X_train = X_train.reshape(-1, 1)
X_test = X_test.reshape(-1, 1)

y_train = y_train.reshape(-1, 1)
y_test = y_test.reshape(-1, 1)

print("X_train :", X_train.shape)
print("y_train :", y_train.shape)
print("X_test  :", X_test.shape)
print("y_test  :", y_test.shape)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred_train = model.predict(X_train)
mse_train = mean_squared_error(y_pred_train,y_train)

print("Erreur quadratique moyenne :", mse_train)
```

```python
y_pred_test = model.predict(X_test)
mse_test = mean_squared_error(y_test,y_pred_test)

print("Erreur quadratique moyenne :", mse_test)
```

```python
import matplotlib.pyplot as plt

train_count = len(X_train)
test_count = len(X_test)

labels = ['Train Data', 'Test Data']
counts = [train_count, test_count]

plt.bar(labels, counts, color=['blue', 'red'])
plt.title('Count of Train and Test Data')
plt.ylabel('Count')
plt.show()
```
